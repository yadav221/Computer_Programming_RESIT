```python
#Task_05
# Initialize variables
count = 0  # number of strings
longest = 0  # length of longest string

while True:
    # Ask user to enter string
    text = input("Enter a string (press Enter to stop): ")
    
    # Stop if empty input
    if text == "":
        break  # exit loop
    
    # Increase count
    count += 1
    
    # Check length of current string
    if len(text) > longest:
        longest = len(text)  # update longest length

# Display results
print("Total strings entered:", count)
print("Length of longest string:", longest)
```

```python
#Task_04
# Function to check valid 5-letter lowercase word
def is_five_letter_word(word):
    # Check length is 5
    if len(word) != 5:
        return False  # not valid if not 5 letters
    
    # Check each character
    for char in word:
        if not char.islower():  # check lowercase letter
            return False  # return False if any char is not lowercase
    
    return True  # return True if all checks pass

# Example test
print(is_five_letter_word("apple"))  # True
print(is_five_letter_word("App1e"))  # False
```

```python
#Task_03
# Ask user to enter number of students
students = int(input("Enter number of students: "))  # convert input to integer

# Calculate number of teams (each team has 11 players)
teams = students // 11  # integer division gives full teams

# Calculate remaining students
leftover = students % 11  # remainder gives students without team

# Display results
print("Number of teams:", teams)  # show teams
print("Students without team:", leftover)  # show leftover
```
